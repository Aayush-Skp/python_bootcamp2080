## Scatter Graph

```python
plt.scatter(mydata['SepalLengthCm'],mydata['SepalWidthCm'])
plt.xlabel('SepalLengthCm')
plt.ylabel('SepalWidthCm')
```

```python
plt.figure(figsize=(8, 6))  # Create a scatter plot

colors = {'Iris-setosa': 'red', 'Iris-versicolor': 'green', 'Iris-virginica': 'blue'}  # Define colors for each species

# Plot each species separately
for species, color in colors.items():  
    """
    for species, color in colors.items(): is a for loop that iterates over the items in
    the colors dictionary. In each iteration, it assigns the key (species) and value (color) 
    from the dictionary to the variables species and color respectively.
    """
    species_data = mydata[mydata['Species'] == species]
    """
    species_data = mydata[mydata['Species'] == species]: This line filters the data in mydata DataFrame
    based on the species. It creates a new DataFrame called species_data containing only the rows where
    the 'Species' column matches the current species in the iteration.
    """
    
    plt.scatter(species_data['SepalLengthCm'], species_data['SepalWidthCm'], c=color, label=species)
    """
    plt.scatter(...): This function creates a scatter plot. It takes the x-axis values (species_data['SepalLengthCm']) 
    and the y-axis values (species_data['SepalWidthCm']) from species_data. These values represent the sepal 
    length and sepal width of the iris flowers of the current species.
    """
    
    
plt.xlabel('Sepal length (cm)')
plt.ylabel('Sepal width (cm)')
plt.title('Scatter plot of Iris dataset')
plt.legend() 
plt.show()
```

<b>plt.legend():</b> This command adds a legend to the plot.
The legend provides a key to help identify which colors correspond to which species in the scatter plot. It's essential when there are multiple datasets or groups represented in the plot.

### Making Pie Chart

```python
# Count occurrences of each class
class_counts = mydata['Species'].value_counts()

# Plotting the pie chart
explode = (0.1, 0,0)
colors= ['pink','yellow','purple']
plt.figure(figsize=(10, 10))
plt.pie(class_counts, labels=class_counts.index, autopct='%1.1f%%', startangle=90, explode=explode, colors=colors, shadow=True, 
       )
plt.title('Distribution of Iris Dataset')
#plt.legend()
plt.show()
#print(class_counts.index)
```

The <b>autopct</b> parameter in the plt.pie() function of Matplotlib stands for "automatic percent" or "auto percentage." It is used to format the numeric values displayed as percentages on the pie chart.

```python
#Compulsory parameters for pue chart
sizes = [10, 20, 30, 40]  # Sizes of the wedges
labels = ['A', 'B', 'C', 'D']  # Labels for the wedges

plt.pie(sizes, labels=labels)
plt.show()
```
